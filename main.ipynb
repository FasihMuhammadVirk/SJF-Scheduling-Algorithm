```python
Print("Shortest Job First")
print("")

#Input
jobs = []
bursttime =[]
arrivaltime = []

jj = int(input('Enter How Many Jobs are there e.g(5): '))

for i in range(0,jj):
  jobs.append(f'P{i+1}')


for i in range(0,jj):

  print("")  
  aa = int(input(f"Enter the arrivail time of the {i+1} Job: "))
  arrivaltime.append(aa)
  bb = int(input(f"Enter the  burst time   of the {i+1} Job: "))
  bursttime.append(bb)

# Zipped the data
zipped = zip(jobs, bursttime,arrivaltime)


# Converting to list
zipped = list(zipped)
  
  
# Using sorted and lambda
res = sorted(zipped, key = lambda x: x[2] and x[1])

jobs = []

for i in res:
  jobs.append(i[0])

#Completetion Time 
completetiontime = []

szero = 0 
sone = 1

for j,b,a in res:

  if 0 in arrivaltime:
    szero += b
    completetiontime.append(szero)
  else:
    sone += b
    completetiontime.append(sone)


#Turn Around Time
tat = []
a = 0
for i,j,k in res:
  tat.append( completetiontime[a] - k )
  a += 1


#Waiting Time
wt = []
w = 0
for i,j,k in res:
  wt.append( tat[w] - j )
  w += 1


res2 = list(zip(completetiontime,tat,wt))

final = sorted(list(zip(res,res2)),key = lambda x: x[0][0] )

#Average TAT
avgtat = str(sum(tat)/jj)


#Average WT
avgwt = str(sum(wt)/jj)


#Output
print('Tabel')
print(' ')

print('Jobs   Burst Time    Arrival Time    CT        TAT        WT    ')
for a,b in final :
  print('{}          {}              {}         {}         {}        {}'.format(a[0],a[1],a[2],b[0],b[1],b[2]))


print(' ')
print(f'The Average TAT is {avgtat}')
print(f'The Average WT is {avgwt}')
```
